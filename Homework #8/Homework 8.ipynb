```python
#Arman Khondker 
#EE 313

import numpy as np
from matplotlib import pyplot as plt

j = (-1)**1/2
omega = np.logspace(-2,4)
H = (5j*omega)/(1+25*j*omega-50*omega**2)    #When you plug in R = 5, C = 0.1, L= 0.2 
plt.semilogx(omega, 10*np.log10(np.abs(H)))
plt.title('Gain of absolute value of H(jw)')
plt.xlabel('Frequency')
plt.ylabel('Magnitude (in dB)')
plt.show()


```
