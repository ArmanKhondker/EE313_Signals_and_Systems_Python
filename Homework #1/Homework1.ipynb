```python
import numpy as np
from matplotlib import pyplot as plt

dt = 1/1000
t = np.arange(-1,1,dt)
f0 = 4
x = 100 * np.real(np.exp(1j*(2*np.pi*f0*(t - 0.75))))

plt.subplot(3,1,1)

plt.plot(t,x)
plt.title('section of a sinusoid')
plt.xlabel('time (sec)')
plt.show()
```

```python
dt = 1/1000
t = np.arange(-1,1,dt)
f0 = 40
y = 10 * np.real(np.exp(1j*(2*np.pi*f0*(t - 0.75))))

plt.subplot(3,1,1)

plt.plot(t,y)
plt.title('section of a sinusoid')
plt.xlabel('time (sec)')
plt.show()
```

```python
z=x*y

plt.subplot(3,1,1)

plt.plot(t,z)
plt.title('section of a sinusoid')
plt.xlabel('time (sec)')
plt.show()
```
