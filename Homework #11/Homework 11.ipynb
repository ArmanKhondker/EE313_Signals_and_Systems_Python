```python
#Arman Khondker
#EE 313 - Homework 11
import numpy as np
from scipy import signal, misc
from matplotlib import pyplot as plt
y = [1, -0.6, -0.35, 0.9]
x = [1, -1, 0.49, -0.49]
```

```python
# The pole zero plot
a = signal.lti(y , x)
w = np.linspace(0,2*np.pi,101)
plt.plot(np.cos(w), np.sin(w), 'b')
plt.plot(np.real(a.zeros), np.imag(a.zeros), 'ko', markerfacecolor = 'b')
plt.plot(np.real(a.poles), np.imag(a.poles), 'kx')
plt.axes().set_aspect('equal')
plt.title('Pole-zero Plot')
plt.show()

w, h = signal.freqz(y,x)
plt.subplot(2,1,1)
plt.plot(w,20*np.log10(np.real(abs(h))))
plt.xlabel('Frequency in rad/sample')
plt.ylabel('Magnitutde in dB')
plt.subplot(2,1,2)
plt.plot(w, np.angle(h))
plt.ylabel('Phase in radians')
plt.xlabel('Frequency in rad/sample')
plt.show()
```
