```python
#Arman Khondker

import numpy as np 
import matplotlib.pyplot as plt


n = np.linspace(-4,4,9)
yn = ((np.cos(np.pi*n/2)-np.sin(np.pi*n/2)+np.cos(np.pi*n)+np.pi/2)/(np.pi*n/2))
plt.stem(n,yn)
plt.xlabel('k')
plt.ylabel('Ak')
plt.show()

yn2 = (((np.cos(np.pi*n/3))-np.sin(np.pi*n/3)+np.cos(np.pi*5/3)+np.sin(np.pi*n/3)/(np.pi/3)))

plt.stem(n,yn2)
plt.xlabel('k')
plt.ylabel('Ak')
plt.suptitle('b)')
plt.show()
```
